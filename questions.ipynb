Here is the code used to plot the price/quantities function:


```julia
using PyPlot
#Setting the demand function
function demand(p)
  return 2.0.*p.^(-0.5)
end
prices=linspace(0,5,200)
dem=demand(prices)
figure(1)
plot(prices,dem, color="blue")
plot(prices, ones(200).*2, color="red")
plot(prices, ones(200), color="red")
title("Equilibria for p=1 and p=4" )
xlabel("Prices")
ylabel("Quantities")
```

Analytically, the change in consummer surplus (S) is computed as:
$$\Delta S(p_1,p_2)=S(p_1)-S(p_2)$$
which is:
$$\Delta S(1,4)= S(1)-S(4)$$
$$\Delta S(1,4)= 2-\int_1^\infty 2p^{−0.5} dp - \left(4 - \int_4^\infty 2p^{−0.5} dp \right) $$
$$\Delta S(1,4)= -2 + \int_1^4 2p^{−0.5} dp $$
$$\Delta S(1,4)= -2 + \left[4 \times p^{0.5} \right]_1^4$$
$$\Delta S(1,4)= -2 + 8 - 4 = 2$$
